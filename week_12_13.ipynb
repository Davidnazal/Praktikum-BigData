```python
# Week 12
import pandas as pd

data = {
    'Name' : ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age' : [24, 30, None, 22, 35],
    'Salary' : [48000,  None, 57000, None, 60000]
}
df = pd.DataFrame(data)

df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df.dropna(subset=['Name'], inplace=True)
print('After cleaning:\n',df)
                
```

```python
# Exercise 1
import pandas as pd
df = pd.read_csv("/home/hadoop/Downloads/tested.csv")
print("Misiing values per column:\n", df.isnull().sum(),'\n')

df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df['Cabin'].fillna(df['Cabin'].mode(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode(), inplace=True)

df.dropna(subset=['Name'], inplace=True)

print("Afteer cleaning:\n", df.head())

      
```

```python
# Exercise 2
df['Sex'] = df['Sex'].str.capitalize()
df['Embarked'] = df['Embarked'].str.upper()
print("Standarized categorical:\n", df[['Sex', 'Embarked']].head())
```

```python
# Exerciise 3
import numpy as np
from sklearn.preprocessing import MinMaxScaler

df['Fare_log'] = np.log1p(df['Fare'])

scaler = MinMaxScaler()
df[['Age_norm', 'Fare_norm']] = scaler.fit_transform(df[['Age', 'Fare']])

print("Transformedd data:\n", df[['Age', 'Fare', 'Age_norm', 'Fare_norm', 'Fare_log']].head())
```

```python
# Exercise 4
filtered_df = df[(df['Sex'] == 'Female') & (df['Age'] > 20)]

df_no_dup = df.drop_duplicates()

print("Filtered data:\n", filtered_df.head())
print("\nTotal duplicates droped:", len(df) - len(df_no_dup))
```

```python
# Exercise 5
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Fare'] < lower) | (df['Fare'] > upper)]
print("Jumlah outlier Fare:", len(outliers), "\n")

df['Fare_winsor'] = df['Fare'].clip(lower, upper)
print("Fare after outlier handling:|n", df[['Fare', 'Fare_winsor']].head())
```

```python
# Weeek 13
# Berdasarkan jenis kelamin
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("/home/hadoop/Downloads/tested.csv")

grouped = df.groupby(['Sex', 'Survived']).size().unstack()

plt.figure(figsize=(8,5))
grouped.plot(kind='bar', figsize=(8,5))

plt.title("Perbandingan Penumpang Selamat VS Tidak Selamat")
plt.xlabel("Jenis kelamin")
plt.ylabel("Jumlah penumpang")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
                      
```

```python
# Berdasarkan status "Selamat" dengan seaborn

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='coolwarm')

plt.title("Scatter plot: Usia VS Tiket")
plt.tight_layout()
plt.show()
```

```python
survival_by_sex  = df.groupby('Sex')['Survived'].mean().reset_index()

plt.figure(figsize=(6,5))
sns.scatterplot(data=survival_by_sex, x='Sex', y='Survived', s=200, color='orange')

plt.title("Persentase selamat berdasarlan jenis kelamin")
plt.ylim(-0.1,1.1)
plt.tight_layout()
plt.show()
```

```python
survival_by_class  = df.groupby('Pclass')['Survived'].mean().reset_index()

plt.figure(figsize=(6,5))
sns.scatterplot(data=survival_by_class, x='Pclass', y='Survived', s=200, color='orange')

plt.title("Persentase selamat berdasarkan Tiket")
sns.regplot(x='Pclass', y='Survived', data=df)
plt.ylim(0,1)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(df[['Age', 'Fare', 'Pclass', 'Survived']].corr(),
            annot=True, cmap='coolwarm', linewidths=0.5)

plt.title("Heatmap korelasi variable numerik")
plt.tight_layout()
plt.show()
```
